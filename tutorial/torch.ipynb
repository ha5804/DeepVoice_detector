# 1.DataLoader

torch에서 제공하는 DataLoader는 Dataset을 학습에 쓰기 좋은 형태로 변환해주는 클래스이다.

### batch size
- batch size는 한 번에 모델에게 넘겨줄 데이터 갯수이다.
batch size가 1이라면 아래 예제 기준으로 샘플 1개 (1,2,3)만 넘겨준다.
따라서 총 2번의 연산 진행.

### shuffle
- 매 epoch마다 데이터 순서를 랜덤으로 섞는다. 항상 같은 순서로 학습하면, 모델이 순서를 외워버리는 문제가 발생한다.

### drop_lst
- 마지막 batch가 안맞으면 버릴지 결정한다. 예를들어 데이터 5개, batchsize = 2라면, batch 1 = (1,2), batch 2 = (3, 4) , batch 3 = (5)로 1개가 남는 상황이 발생한다. 이때 마지막 3번째 배치를 버릴지 결정한다.

### num_workers
- 데이터를 불러오는 프로세스 수로, 프로세스를 병렬로 로딩할지 결정한다. 작은 데이터에서는 0, 실무에서는 2, 4, 8과 같은 수를 사용한다.

```python
from torch.utils.data import DataLoader

train_dataset = [[1,2,3],[4,5,6]]

train_loader = DataLoader(
    train_dataset,
    batch_size=1,
    shuffle=True,
    drop_last=False,
    num_workers = 0
)
```
